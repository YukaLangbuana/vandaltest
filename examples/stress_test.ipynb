```python
import os
import sys
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)

import mayhem
```

#### Define Url, Payload, and Headers

```python
url = "http://34.101.111.158:8070/api/v1/auth/login"

payload = {
  "username": "anggaaditya",
  "password": "Shalawa18",
  "rememberMe": False,
  "grant_type": "password"
}
headers = {
  'X-XSRF-TOKEN': '66ccf0b2-6ebb-4837-a567-50bd77a6110a',
  'Content-Type': 'application/json',
  'Cookie': 'XSRF-TOKEN=66ccf0b2-6ebb-4837-a567-50bd77a6110a'
}
```

#### Run Stress Test

```python
result = mayhem.stress_test("POST", url, data=payload, headers=headers, vus=100, duration=60)
```
